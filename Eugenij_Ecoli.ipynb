```python
import pandas as pd

# download our data
coords = pd.read_csv("/content/out.coords", sep='\t', comment='#', header=None)
```

```python
coords.head()
```

```python
# Defining column names based on the structure analyzed
column_names = [
    "Start1", "End1", "Start2", "End2", "Length1", "Length2",
    "% Identity", "Len Ref", "Len Query", "Cov Ref", "Cov Query",
    "Ref Seq Name", "Query Seq Name"
]

# Assigning the column names
coords.columns = column_names

# Saving the file back with headers
output_file_path = "/content/out_with_headers.coords"
coords.to_csv(output_file_path, sep='\t', index=False)

output_file_path
```

```python
coords.head()
```

```python
# create new dataframe with % Identity < 90%
threshold = 90
low_identity = coords[coords["% Identity"] < threshold]
nodes_list = low_identity['Query Seq Name'].unique()
df_nodes = pd.DataFrame(nodes_list, columns=["Query Seq Name"])

#low_identity.head()
#nodes_list

# saving with 90% treshhold only unique NODs
output_file_path = "/content/selected.txt"
df_nodes.to_csv(output_file_path, header=False, index=False)
df_nodes
```
