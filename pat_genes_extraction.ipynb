```python
import pandas as pd

# download our data
blastp = pd.read_csv("/content/blastp_output.txt", sep='|', comment='#', header=None)
```

```python
blastp
```

```python
# Defining column names based on the structure analyzed
column_names = ["type", "number", "gene's name"
]

# Assigning the column names
blastp.columns = column_names
```

```python
# string's filtration with word 'shigella'
shigella_genes = blastp[blastp["gene's name"].str.contains('Shigella', case=False)]

# searching unique names
unique_genes = shigella_genes["number"].unique()

# file's saving
with open('shigella_genes.txt', 'w') as file:
    for gene in unique_genes:
        file.write(gene + '\n')

print(f'Сохранено {len(unique_genes)} уникальных генов в файл shigella_genes.txt')
```

```python
shigella_genes_unique = shigella_genes["gene's name"]
shigella_genes_unique
```

```python
# string's filtration with word 'stx'
stx_genes = blastp[blastp["gene's name"].str.contains('stx', case=False)]

# searching unique names
unique_genes = stx_genes["number"].unique()

# file's saving
with open('stx.txt', 'w') as file:
    for gene in unique_genes:
        file.write(gene + '\n')

print(f'Сохранено {len(unique_genes)} уникальных генов в файл stx.txt')
stx_genes
```

```python
# Filter rows in 'blastp' DataFrame where "gene's name" contains both 'stx' and 'phage' (case insensitive)
phage_genes = blastp[blastp["gene's name"].str.contains('Shigella', case=False) & blastp["gene's name"].str.contains('phage', case=False)]

# Extract unique 'number' values
unique_phage_genes = phage_genes['number'].unique()

# Save unique numbers to file
with open('stx.txt', 'w') as file:
    for gene in unique_phage_genes:
        file.write(gene + '\n')

print(f'Saved {len(unique_phage_genes)} unique genes to file stx.txt')

# Display the DataFrame with the filtered genes
phage_genes
```
